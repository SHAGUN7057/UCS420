```python
#ques1
import numpy as np
gfg=np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
#1.i)
print("sum of matrix",np.sum(gfg))
#1.ii)
print("sum of matrix row wise",np.sum(gfg,axis=1))
#1.iii)
print("sum of matrix col wise",np.sum(gfg,axis=0))
```

```python
#ques2
#2.a)
array = np.array([10, 52, 62, 16, 16, 54, 453])
#a.i)
arr1=np.sort(array)
print("sorted array",arr1)
#a.ii)
print("Indices of sorted array",np.argsort(array))
#a.iii)
print("4 smallest elements",arr1[:4])
#a.iv)
print("5 largest elements",arr1[-5:])
```

```python
#2.b)
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
#b.i)
print("Integer elements only",array[array == array.astype(int)])
#b.ii)
print("Float elements only",array[array != array.astype(int)])
```

```python
#ques3
#a)
X=ord('S')+ord('D')
sales=np.array([X,X+50,X+100,X+150,X+200])
print("sales array",sales)
#b)
tax_rate=((X%5)+5)/100
sales=sales*(1+tax_rate)
print("sales with taxes",sales)
#c)
discount=np.where(sales<X+100,sales*0.95,sales*0.90)
print("sales after discount",discount)
#d)
sales_mat=np.tile(sales,(3,1))
print("three weeks of sales \n",sales_mat)
#e)
weeks = np.arange(3).reshape(3, 1)  
growth_factor = 1 + (weeks * 0.02) 
final_sales = sales * growth_factor
print("final sales in 3 weeks \n",final_sales)
```

```python
#ques4 
import numpy as np 
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100) 
y_square = x**2
y_sin = np.sin(x)
y_exp = np.exp(x)
y_log = np.log(np.abs(x) + 1)

plt.subplot(2, 2, 1)
plt.plot(x, y_square )

plt.subplot(2, 2, 2) 
plt.plot(x, y_sin)

plt.subplot(2, 2, 3) 
plt.plot(x, y_exp)

plt.subplot(2, 2, 4)
plt.plot(x, y_log)

plt.show()
```
