```python
#ques1)
import numpy as np
import matplotlib.pyplot as plt
M=float(input("enter desired no"))
x=np.linspace(-10,10,100)
y1=M*x*x
y2=M*np.sin(x)
plt.plot(x,y1,'r--',label=f'y = {M} * x²')
plt.plot(x,y2,'b-',label=f'y = {M} * sin(x)')
plt.title(f'Plot of y= {M} * x^2 and plot of y={M}* sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#ques 2)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {'Subject': ['Math', 'Science', 'English', 'History', 'Art'],
        'Score': [85, 90, 78, 88, 95]}
df = pd.DataFrame(data)
sns.set_style("whitegrid")
plt.figure(figsize=(8, 6))
ax = sns.barplot(x='Subject', y='Score', data=df,  hue='Subject', palette='husl', legend=False)

for i, score in enumerate(df['Score']):
    ax.text(i, score + 1, str(score), ha='center', fontsize=12, fontweight='bold')
plt.title('Student Scores by Subject', fontsize=14)
plt.xlabel('Subjects', fontsize=12)
plt.ylabel('Scores', fontsize=12)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.show()
```

```python
#ques3)
roll=102317057
np.random.seed(roll)

data=np.random.randn(50)
y_sum = np.cumsum(data)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].plot(y_sum, marker='o', linestyle='-', color='b')
axs[0, 0].set_title('Cumulative Sum Line Plot')
axs[0, 0].set_xlabel('Index')
axs[0, 0].set_ylabel('Cumulative Sum')
axs[0, 0].grid(True)

for i in range(2):
    for j in range(2):
        if (i, j) != (0, 0):
            axs[i, j].axis('off')

plt.show()

noise_x = np.random.randn(50) * 0.1
noise_y = np.random.randn(50) * 0.1

plt.figure(figsize=(6, 4))
plt.scatter(np.arange(50) + noise_x, data + noise_y, color='b', alpha=0.7)
plt.title("Scatter Plot with Random Noise", fontsize=14, fontweight='bold')
plt.xlabel("Index", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
 
data = pd.read_csv("sales.csv")

print("Column names in dataset:", data.columns)

product_columns = ["facecream", "facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"]

plt.figure(figsize=(10, 5))
sns.lineplot(x=data["month_number"], y=data["total_profit"], marker="o", color="b", label="Total Profit")
plt.title("Total Profit Over Months")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
for product in product_columns:
    sns.lineplot(x=data["month_number"], y=data[product], label=product, marker="o")

plt.title("Product Sales Over Months")
plt.xlabel("Month Number")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()


plt.figure(figsize=(12, 6))
data[product_columns].sum().plot(kind="bar", color=["r", "g", "b", "y", "c", "m"])
plt.title("Total Sales of Each Product")
plt.xlabel("Products")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```
