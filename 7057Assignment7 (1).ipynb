```python
#ques1)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#PART I
roll_no = 102317057
np.random.seed(roll_no)
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
columns = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]
sales = np.random.randint(1000, 5001, size=(12, 4))
sales_df = pd.DataFrame(sales, columns=columns, index=months)
```

```python
#PART II
print("First 5 rows of Sales Data:")
print(sales_df.head())
print("\nSummary Statistics:")
print(sales_df.describe())
sales_df["Total Sales"] = sales_df.sum(axis=1)
category_totals = sales_df.sum()

print("\nTotal Sales Per Category:")
print(category_totals[:-1])

print("\nTotal Sales Per Month:")
print(sales_df["Total Sales"])

sales_growth = sales_df.diff()
average_growth = sales_growth.mean()

print("\nAverage Sales Growth Per Category:")
print(average_growth[:-1]) 

sales_df["Growth Rate"] = sales_df["Total Sales"].pct_change() * 100

roll_number = 102317057
if roll_number % 2 == 0:
    sales_df["Electronics"] *= 0.90  
else:
    sales_df["Clothing"] *= 0.85  


print("\nUpdated Sales Data with Growth Rate and Discounts:")
print(sales_df)
```

```python
#Part III
plt.figure(figsize=(10, 6))
sales_df.drop(columns=["Total Sales", "Growth Rate"]).plot(kind='line', marker='o', title='Monthly Sales Trends', figsize=(10,8))
plt.xlabel("Month")
plt.ylabel("Sales Units")
plt.legend(title="Product Categories")
plt.show()

plt.figure(figsize=(10, 6))
sales_df.drop(columns=["Total Sales", "Growth Rate"]).plot(kind='box', title='Sales Distribution by Category')
plt.ylabel("Sales Units")
plt.show()
```

```python
#ques 2)
import numpy as np

array = np.array([[1, -2, 3], [-4, 5, -6]])

absolute_values = np.abs(array)
print("Element-wise Absolute Values:\n", absolute_values)

flattened_percentiles = np.percentile(array, [25, 50, 75])
print("\nPercentiles of Flattened Array (25th, 50th, 75th):\n", flattened_percentiles)

column_percentiles = np.percentile(array, [25, 50, 75], axis=0)
print("\nPercentiles for Each Column (25th, 50th, 75th):\n", column_percentiles)

row_percentiles = np.percentile(array, [25, 50, 75], axis=1)
print("\nPercentiles for Each Row (25th, 50th, 75th):\n", row_percentiles)

mean_flat = np.mean(array)
median_flat = np.median(array)
std_flat = np.std(array)

print("\nMean, Median, and Std Dev of Flattened Array:")
print("Mean:", mean_flat, "Median:", median_flat, "Std Dev:", std_flat)

mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)

print("\nMean, Median, and Std Dev for Each Column:")
print("Mean:", mean_col, "\nMedian:", median_col, "\nStd Dev:", std_col)

mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)

print("\nMean, Median, and Std Dev for Each Row:")
print("Mean:", mean_row, "\nMedian:", median_row, "\nStd Dev:", std_row)
```

```python
#ques 3)
import numpy as np

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floor_values = np.floor(a)
ceiling_values = np.ceil(a)
truncated_values = np.trunc(a)
rounded_values = np.round(a)

print("Original Array:\n", a)
print("\nFloor Values:\n", floor_values)
print("\nCeiling Values:\n", ceiling_values)
print("\nTruncated Values:\n", truncated_values)
print("\nRounded Values:\n", rounded_values)
```

```python
#ques 4)
def swap_ele(l,n1,n2):
    temp=l[n1]
    l[n1]=l[n2]
    l[n2]=temp
    return l

list1=[89,65,78,45,98,23,67]
print("Original list",list1)
new_l=swap_ele(list1,3,4)
print("after swap",new_l)

```

```python
def swap_ele(s,n1,n2):
    l=list(s)
    temp=l[n1]
    l[n1]=l[n2]
    l[n2]=temp
    return set(l)   
s1={45,56,67,78,89}
print("original set",s1)
s2=swap_ele(s1,1,3)
print("after swap",s2)
```
